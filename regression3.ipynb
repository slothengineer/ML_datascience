## EDA and feature engineering of algerian forest fire data 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
```

```python
df=pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header=1)
```

```python
df.head()
```

```python
df.info()
```

```python
##missing value
df[df.isnull().any(axis=1)]
```

```python
df.shape
```

```python
df.loc[:122,'Region']=0
df.loc[122:,'Region']=1
```

```python
df.info()
```

```python
df[['Region']]=df[['Region']].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
## removing null values
df=df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
df.columns=df.columns.str.strip()
```

```python
df.columns
```

```python
df[['day','month','year','Temperature','RH','Ws']]=df[['day','month','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

```python
objects=[features for features in df.columns if df[features].dtype=='O']
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
objects
```

```python
df.describe()
```

```python
df.to_csv("Algerian_forest_fires_dataset_cleaned.csv",index=False)
```

```python
df_c=df.drop(['day','month','year'],axis=1)
```

```python
df_c['Classes']=np.where(df_c['Classes'].str.contains('not fire'),0,1)
```

```python
plt.style.use('seaborn')
df_c.hist(figsize=(20,15),bins=50)
plt.show()
```

```python
## percentage for pie chart
percentage=df_c['Classes'].value_counts(normalize=True)*100
```

```python
classlables=['fire','not fire']
plt.figure(figsize=(12,8))
plt.pie(percentage,labels=classlables,autopct='%1.1f%%')
plt.title("pie chart of classes")
plt.show()
```
