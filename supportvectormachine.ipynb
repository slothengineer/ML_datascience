## Support vector classifier

```python
import pandas as pd
import seaborn as sb
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_samples=1000,n_features=2,n_classes=2,n_clusters_per_class=2,n_redundant=0)
```

```python
X
```

```python
y
```

```python
x=pd.DataFrame(X)[0]
```

```python
sb.scatterplot(x=pd.DataFrame(X)[0],y=pd.DataFrame(X)[1],hue=y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=10)
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC(kernel='linear')
```

```python
svc.fit(X_train,y_train)
```

```python
svc.coef_
```

```python
y_pred=svc.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(accuracy_score(y_test,y_pred))
```

### hyperparameter tuning

```python
from sklearn.model_selection import GridSearchCV 
```

```python
parameters={
            'C':[0.1,1,10,100,1000],
            'gamma':[1,0.1,0.01,0.001,0.0001],
            'kernel':['linear']
}
```

```python
grid=GridSearchCV(SVC(),param_grid=parameters,refit=True,cv=5,verbose=3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred1=grid.predict(X_test)
```

```python
print(accuracy_score(y_test,y_pred1))
```

```python
print(confusion_matrix(y_test,y_pred1))
```

```python
print(classification_report(y_test,y_pred1))
```

## svm kernels

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(-5.0,5.0,100)
y=np.sqrt(10**2-x**2)
y=np.hstack([y,-y])
x=np.hstack([x,-x])
```

```python
x1=np.linspace(-5.0,5.0,100)
y1=np.sqrt(5**2-x1**2)
y1=np.hstack([y1,-y1])
x1=np.hstack([x1,-x1])
```

```python
plt.scatter(y,x)
plt.scatter(y1,x1)
```

```python
import pandas as pd
df1=pd.DataFrame(np.vstack([x,y]).T,columns=['X1','X2'])
df1['Y']=0
df2=pd.DataFrame(np.vstack([x1,y1]).T,columns=['X1','X2'])
df2['Y']=1
df=df1.append(df2)
df.head()
```

### polynomial transformation

```python
df['X1_square']=df['X1']**2
df['X2_square']=df['X2']**2
df['X1*X2']=df['X1']*df['X2']
df.head()
```

```python
!pip install plotly
```

```python
import plotly.express as pb
fig=pb.scatter_3d(df,x='X1_square',y='X2_square',z='X1*X2',color='Y')
fig.show()
plt.show()
```

```python
X=df[['X1_square','X2_square','X1*X2']]
y=df['Y']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
X_train.head()
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
svc=SVC(kernel='linear')
svc.fit(X_train,y_train)
y_pred=svc.predict(X_test)
print(accuracy_score(y_test,y_pred))
```
