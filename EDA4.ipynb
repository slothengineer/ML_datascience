```python
df_copy=df.copy()
```

```python
df_copy=df_copy.drop(df_copy.index[10472])
```

```python
df_copy[~df_copy['Reviews'].str.isnumeric()]
```

```python
df_copy['Reviews']=df_copy['Reviews'].astype(int)
```

```python
df_copy.info()
```

```python
df_copy['Size'].unique()
```

```python
df_copy['Size'].isnull().sum()
```

```python
df_copy['Size']=df_copy['Size'].str.replace('M','000')
df_copy['Size']=df_copy['Size'].str.replace('k','')
df_copy['Size']=df_copy['Size'].replace('Varies with device',np.nan)
df_copy['Size']=df_copy['Size'].astype(float)
```

```python
df_copy.info()
```

```python
df_copy['Size']
```

```python
df_copy['Size'].unique()
```

```python
df_copy['Size'].mean()
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Price'].unique()
```

```python
chars_to_remove=['+',',','$']
cols_to_clean=['Installs','Price']
for items in chars_to_remove:
    for cols in cols_to_clean:
        df_copy[cols]=df_copy[cols].str.replace(items,'')
```

```python
df_copy['Price'].unique()
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Installs']=df_copy['Installs'].astype(int)
df_copy['Price']=df_copy['Price'].astype(float)
```

```python
df_copy.info()
```

```python
df_copy['Last Updated'].unique()
```

```python
df_copy['Last Updated']=pd.to_datetime(df_copy['Last Updated'])
```

```python
df_copy['Last Updated']
```

```python
df_copy['Day']=df_copy['Last Updated'].dt.day
```

```python
df_copy['Month']=df_copy['Last Updated'].dt.month
```

```python
df_copy['Year']=df_copy['Last Updated'].dt.year
```

```python
df_copy.info()
```

```python
df_copy.to_csv('google-cleaned.csv')
```

## EDA

```python
df_copy.head()
```

```python
df_copy[df_copy.duplicated('App')].shape
```

```python
df_copy=df_copy.drop_duplicates(subset=['App'],keep='first')
```

```python
df_copy.shape
```

```python
## data exploration
num_fea=[feature for feature in df_copy.columns if df_copy[feature].dtype!='O']
cat_fea=[feature for feature in df_copy.columns if df_copy[feature].dtype=='O']
```

```python
print('we have {} num_fea: {}'.format(len(num_fea),num_fea))
print('\nwe have {} cat_fea: {}'.format(len(cat_fea),cat_fea))
```

```python
## proportion of count data on categorical columns
for col in cat_fea:
    print(df_copy[col].value_counts(normalize=True)*100)
    print('----------------------------------------------------------------')
```

```python
## proportion of count data on numerical columns
plt.figure(figsize=(15,15))
plt.suptitle('univariate analysis of numerical features',fontsize=25,fontweight='bold',alpha=0.8,y=1.)

for i in range(0,len(num_fea)):
    plt.subplot(5,3 , i+1)
    sb.kdeplot(x=df_copy[num_fea[i]],shade=True,color='b')
    plt.xlabel(num_fea[i])
    plt.tight_layout()
    
```

```python
plt.figure(figsize=(20,20))
plt.suptitle('univariate analysis of categorical features',fontsize=25,fontweight='bold',alpha=0.8,y=1.)
cat=['Type','Content Rating']
for i in range(0,len(cat)):
    plt.subplot(2,2 , i+1)
    sb.countplot(x=df_copy[cat[i]],palette='Set2')
    plt.xlabel(cat[i])
    plt.xticks(rotation=45)
    plt.tight_layout()
    
```
