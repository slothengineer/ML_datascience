## DBScan clustering anomaly detection

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_circles
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X,y=make_circles(n_samples=750,factor=0.3,noise=0.1)
X
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
dbs=DBSCAN(eps=0.10)
```

```python
dbs.fit_predict(X)
```

```python
dbs.labels_
```

## local outlier factor

```python
from sklearn.datasets import make_moons
```

```python
plt.scatter(X[:,0],X[:,1],c=dbs.labels_)
```

```python
X,y=make_moons(n_samples=1000,noise=0.10)
```

```python
X
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

```python
from sklearn.neighbors import LocalOutlierFactor
```

```python
lof=LocalOutlierFactor(n_neighbors=5,algorithm='auto')
```

```python
predicted=lof.fit_predict(X)
```

```python
lof.negative_outlier_factor_
```

```python
plt.scatter(X[:,0],X[:,1],c=lof.negative_outlier_factor_)
```

```python
lof.contamination
```

```python
lof.effective_metric_
```

## Isolation forest

```python
import pandas as pd
df=pd.read_csv('healthcare.csv')
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df.iloc[:,0],df.iloc[:,1])
```

```python
from sklearn.ensemble import IsolationForest
ifd=IsolationForest(contamination=0.2)
```

```python
ifd.fit(df)
```

```python
predicted=ifd.predict(df)
```

```python
predicted
```

```python
import numpy as np
index=np.where(predicted<0)
```

```python
index
```

```python
X=df.values
```

```python
plt.scatter(df.iloc[:,0],df.iloc[:,1])
plt.scatter(X[index,0],X[index,1],edgecolors='r')
```
